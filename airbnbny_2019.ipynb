```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/Users/carandu/thinkful_projects/capstone_1/AB_NYC_2019.csv')
df.rename(columns = {'neighbourhood_group':'borough'}, inplace = True)
df.head()
```

```python
manhattan = df[df['borough'] == 'Manhattan']
brooklyn = df[df['borough'] == 'Brooklyn']
queens = df[df['borough'] == 'Queens']
bronx = df[df['borough'] == 'Bronx']
staten_island = df[df['borough'] == 'Staten Island']

entire = df[df['room_type'] == 'Entire home/apt']
private = df[df['room_type'] == 'Private room']
shared = df[df['room_type'] == 'Shared room']
```

```python
#determining the top borough
df.borough.value_counts().head(10).plot(kind='bar')
plt.xlabel('Borough')
plt.show()

m = manhattan.id.nunique()
b = brooklyn.id.nunique()
q = queens.id.nunique()
b = bronx.id.nunique()
s = staten_island.id.nunique()

print('The average price for an airbnb in Manhattan in 2019 is:\n{}'.format(manhattan.price.mean()))
print('The average price for an airbnb in Brooklyn in 2019 is:\n{}'.format(brooklyn.price.mean()))
print('The average price for an airbnb in Queens in 2019 is:\n{}'.format(queens.price.mean()))
print('The average price for an airbnb in the Bronx in 2019 is:\n{}'.format(bronx.price.mean()))
print('The average price for an airbnb in Staten Island in 2019 is:\n{}'.format(staten_island.price.mean()))

print(manhattan['price'].max(), manhattan['price'].min())
print(brooklyn['price'].max())
print(queens['price'].max())
print(bronx['price'].max())
print(staten_island['price'].max())

#get the total number of airbnb offerings per borough
boroughs = df.groupby('borough')['id'].nunique()

#get the average days the airbnb is available by borough
availability = df.groupby('borough')['availability_365'].mean()
print(boroughs)

print(availability)
```
