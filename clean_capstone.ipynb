The questions:

Have suicide rates increased over the last 30 years? 
Taking a snapshot in 2015, what are the most suicide-prone countries in the world?
What is the difference between male and female suicide ratios?

The dataset "Suicide Rates Overview 1985 to 2016" by the World Bank compares socio-economic info with suicide rates. Downloaded from Kaggle. It is has 27.8k X 12 columns: country, year, sex, age group, count of suicides, population, suicide rate, country-year composite key, HDI for year, gdp_for_year, gdp_per_capita, generation (based on age grouping average). The population includes info for 15 years old or older, which is the minumum age at which the study looked for its statistics. 

For sake of simplicity I have averaged the suicide rate in years 1995 and 2015 to generate charts on those years.



```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/Users/carandu/thinkful_projects/capstone_1/master.csv')
df = df.drop('HDI for year', 1)
```

```python
#create new dataframes for years 1995 and 2015
df_1995 = df[df['year'] == 1995]
df_2015 = df[df['year'] == 2015]
```

```python

country_1995 = df_1995.groupby('country').apply(np.sum)
suicides_per_100K_1995 = (country_1995['population'] / country_1995['suicides_no']) * 100000
country_1995['suicides_per_100K_1995'] = suicides_per_100K_1995
country_1995[['suicides_no', 'population', 'suicides_per_100K_1995']]
#country_1995.head()
```
